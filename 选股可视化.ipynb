```python
def whatdowebuy(stock_list):
    results = pd.DataFrame()
    data = pro.stock_basic(exchange='', list_status='L', fields='ts_code,symbol,name,area,industry,list_date')
    # 返回所有相关的股票列
    data = data[data['ts_code'].isin(stock_list)]
    list1 = []
    for i in range(0, len(stock_list)):
        try:
            list1.append(data.loc[data['ts_code'] == stock_list[i], 'name'].iloc[0])
        #         list2.append(data.loc[data['industry'] == stock_list[i], 'name'].iloc[0])
        except IndexError:
            list1.append("不上市了")
    #         list2.append("不上市了")
    results["code"] = pd.DataFrame(stock_list)
    results["name"] = pd.DataFrame(list1)
    results = results[~results["name"].isin(["不上市了"])]
    return results["name"]
```
